````python
# Entrada

print("Insira as Informações do Primeito Triângulo")
lado1 = float(input("Digite Aqui o Primeiro Lado: "))
lado2 = float(input("Digite Aqui o Segundo Lado: "))
lado3 = float(input("Digite Aqui o Segundo Operando: "))

print("\n")

print("Insira as Informações do Segundo Triângulo")
lado12 = float(input("Digite Aqui o Primeiro Lado: "))
lado22 = float(input("Digite Aqui o Segundo Lado: "))
lado32 = float(input("Digite Aqui o Segundo Operando: "))

print("\n")

print("Insira as Informações do Terceiro Triângulo")
lado13 = float(input("Digite Aqui o Primeiro Lado: "))
lado23 = float(input("Digite Aqui o Segundo Lado: "))
lado33 = float(input("Digite Aqui o Segundo Operando: "))

print("\n")

# Processamento

## Colocar Inputs para Definir os Valores de Cada Lado dos Três Triângulos
## Utilizar if e elif para Colocar Todos os Tipos (Isóceles, Equilátero, Escaleno) para os Três Triângulos

### Primeiro Triângulo

#### Triangulo Isóceles

if lado1 == lado2 and lado2 == lado3:

# Saída

    print("O Primeiro Triângulo é:")
    print("Isóceles")
    print("\n")

#### Triângulo Equilátero

elif lado1 == lado2 and lado2 != lado3:


# Saída

    print("O Primeiro Triangulo é:")
    print("Equilátero")
    print("\n")

elif lado1 == lado3 and lado3 != lado2:


# Saída

    print("O Primeiro Triangulo é:")
    print("Equilátero")
    print("\n")


elif lado2 == lado3 and lado2 != lado1:

# Saída1
    print("O Primeiro Triangulo é:")
    print("Equilátero")
    print("\n")

#### Triangulo Escaleno

elif 1 != lado2 and lado1 != lado3:
    print("O Primeiro Triangulo é:")
    print("Escaleno")
    print("\n")

else:
    print("Os Valores Estão Incorretos")

### Segundo Triângulo

#### Triangulo Isóceles

if lado12 == lado22 and lado22 == lado32:

# Saída

    print("O Segundo Triângulo é:")
    print("Isóceles")
    print("\n")

#### Triângulo Equilátero

elif lado12 == lado22 and lado22 != lado32:


# Saída

    print("O Segundo Triangulo é:")
    print("Equilátero")
    print("\n")

elif lado12 == lado32 and lado32 != lado22:

# Saída

    print("O Primeiro Triangulo é:")
    print("Equilátero")
    print("\n")


elif lado22 == lado32 and lado22 != lado12:

# Saída1
    print("O Segundo Triangulo é:")
    print("Equilátero")
    print("\n")

#### Triangulo Escaleno

elif lado12 != lado22 and lado12 != lado32:
    print("O Segundo Triangulo é:")
    print("Escaleno")

else:
    print("Os Valores Estão Incorretos")

### Terceiro Triângulo

#### Triangulo Isóceles

if lado13 == lado23 and lado23 == lado33:

# Saída

    print("O Terceiro Triângulo é:")
    print("Isóceles")
    print("\n")

#### Triângulo Equilátero

elif lado13 == lado23 and lado23 != lado33:

# Saída

    print("O Terceiro Triangulo é:")
    print("Equilátero")

elif lado13 == lado33 and lado33 != lado23:

# Saída

    print("O Terceiro Triangulo é:")
    print("Equilátero")
    print("\n")

elif lado23 == lado33 and lado23 != lado12:

# Saída1
    print("O Terceiro Triangulo é:")
    print("Equilátero")
    print("\n")

#### Triangulo Escaleno

elif lado13 != lado23 and lado13 != lado33:
    print("O Terceiro Triangulo é:")
    print("Escaleno")
````
